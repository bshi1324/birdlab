# Bird Lab Question 2

```python
from pandas.plotting import table
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_style("white")
```

```python
bird_counts = pd.read_csv('HistoricalResultsBySpecies.csv')
```

```python
bird_counts = bird_counts.drop(bird_counts.index[45:])
```

```python
bird_counts['Count_yr'] = bird_counts['Count_yr'].astype('int')
bird_counts['Count_yr'] = bird_counts['Count_yr'].apply(lambda x: x+1899)
```

```python
bird_counts.columns = ['Name', 
                       'Species Code', 
                       'Year', 
                       'Number', 
                       'Number/Party Hours', 
                       'Counts Reporting Species', 
                       'Reporting Observers']
```

```python
bird_counts.head()
```

### Question 2.A

```python
x = bird_counts['Year']
y1 = bird_counts['Number']
y2 = bird_counts['Number/Party Hours']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.scatter(x, y1, c='g')
ax2.scatter(x, y2, c='b')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number', color='g')
ax2.set_ylabel('Number/Party Hours', color='b')

plt.title('Number and Number/Party Hours over Year')

plt.savefig('number_numberbypartyhours.png')
plt.show()
```

### Question 2.B

```python
x = bird_counts['Year']
y1 = bird_counts['Number/Party Hours']
```

```python
y2 = bird_counts['Counts Reporting Species']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.scatter(x, y1, c='b')
ax2.scatter(x, y2, c='r')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number/Party Hours', color='b')
ax2.set_ylabel('Counts Reporting Species', color='r')

plt.title('Number/Party Hours and Counts Reporting Species over Year')

plt.savefig('numberbypartyhours_countsreportingspecies.png')
plt.show()
```

```python
y2 = bird_counts['Reporting Observers']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.scatter(x, y1, c='b')
ax2.scatter(x, y2, c='black')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number/Party Hours', color='b')
ax2.set_ylabel('Reporting Observers', color='black')

plt.title('Number/Party Hours and Reporting Observers over Year')

plt.savefig('numberbypartyhours_reportingobservers.png')
plt.show()
```
